#Steps of Analysis

1- Look at the data, do we have NaN, duplicates? 

⚠️ Important here, check for outliers



2- Look at the correlation (multicollienarity) (heatmap for example) between the predictor variables. If we have any variable where the correlation between them is greater than 0.9 we should eliminate one of the two (the one we think is less important in our data).



3- check the normality of the data. In case they are not normal we have to transform them (logarithmic or boxcox transformation).



4- We perform the OLS. If we have variables that have a pvalue greater than 0.05 we eliminate them from our model and run the model again without those variables.
